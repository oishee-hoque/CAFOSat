```python
import os
import json
import numpy as np
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import torch
import pandas as pd
from shapely import wkt
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import torchvision
```

```python


# Mapping dataset type keywords to filenames
DATASET_FILES = {
    'verified':  'cafosat_verified_training.json',
    'augmented': 'cafosat_augmented_training.json',
    'all':       'cafosat_all_training.json',
    'set1':      'cafosat_training_set1.json',
    'set2':      'cafosat_training_set2.json',
    'merged':    'cafosat_merged_training.json'
}

def parse_flag(val):
    if pd.isna(val):
        return 0.0
    if isinstance(val, str):
        val = val.strip().lower()
        if val in ['yes', 'true']:
            return 1.0
        try:
            return 1.0 if float(val) != 0 else 0.0
        except ValueError:
            return 0.0
    if isinstance(val, (int, float)):
        return 1.0 if val != 0 else 0.0
    return 0.0



class CAFOJsonDataset(Dataset):
    def __init__(self, entries, mode='binary', transform=None, image_root='', infra_csv_path=None):
        self.entries = entries
        self.mode = mode
        self.transform = transform
        self.image_root = image_root

        # Load infra info and geometry if mode is 'infra' or 'bbox'
        self.infra_data = None
        if self.mode in ['infra', 'bbox'] and infra_csv_path:
            df = pd.read_csv(infra_csv_path)
            df = df[df['verified_label'].str.lower() != 'not cafo']
            # if 'bbox' in df.columns:
            #     df['bbox'] = df['bbox']
            self.infra_data = df.set_index('patch_file')

    def __len__(self):
        return len(self.entries)

    def __getitem__(self, idx):
        entry = self.entries[idx]

        image_path = entry['image_path']
        full_path = image_path if os.path.isabs(image_path) else os.path.join(self.image_root, image_path)

        image = Image.open(full_path).convert("RGB")
        if self.transform:
            image = self.transform(image)

        label = int(entry.get('label', 0))
        if self.mode == 'binary':
            label = int(label > 0)
        elif self.mode == 'multiclass':
            label = int(label)

        if self.mode == 'bbox' or self.mode == 'all':
            patch_file = os.path.basename(entry['image_path'])

            if int(label) > 0 and self.infra_data is not None and patch_file in self.infra_data.index:
                bbox_val = self.infra_data.loc[patch_file].get('geom_bbox')  # Or 'bbox' if named so

                if isinstance(bbox_val, str):
                    import ast
                    bbox = np.array(ast.literal_eval(bbox_val), dtype=np.float32)
                    original_size = 833
                    resized_size = image.shape[0]
                    scale = resized_size / original_size
                    # Scale the bbox
                    bbox = np.array([coord * scale for coord in bbox])
                    
                elif isinstance(bbox_val, (list, tuple, np.ndarray)):
                    bbox = np.array(bbox_val, dtype=np.float32)
                else:
                    bbox = np.array([5.0, 5.0, 220.0, 220.0], dtype=np.float32)
            else:
                # Fallback for negatives or missing data
                bbox = np.array([5.0, 5.0, 220.0, 220.0], dtype=np.float32)
            if self.mode == 'bbox':
                return image, label, bbox


        if self.mode == 'infra' or self.mode == 'all':
            patch_file = os.path.basename(entry['image_path'])
            if self.infra_data is not None and patch_file in self.infra_data.index:
                infra_row = self.infra_data.loc[patch_file]
                infra_array = np.array([
                    parse_flag(infra_row.get('barn')),
                    parse_flag(infra_row.get('manure_pond')),
                    parse_flag(infra_row.get('grazing_area')),
                    parse_flag(infra_row.get('others'))
                ], dtype=np.float32)
            else:
                infra_array = np.zeros(4, dtype=np.float32)
            if self.mode == 'infra':
                return image, label, infra_array
        if self.mode == 'all':
            return image, label, infra_array, bbox
        else:
            return image, label

def get_cafo_dataloader(dataset_dir, dataset_name='set1', task='binary', split='train',
                         image_root='', batch_size=32, shuffle=True, infra_csv_path=None):
    if dataset_name not in DATASET_FILES:
        raise ValueError(f"Invalid dataset name: {dataset_name}")

    json_path = os.path.join(dataset_dir, DATASET_FILES[dataset_name])
    if not os.path.exists(json_path):
        raise FileNotFoundError(f"JSON file not found: {json_path}")

    with open(json_path, 'r') as f:
        data_dict = json.load(f)

    if split not in data_dict:
        raise ValueError(f"Split '{split}' not found in {json_path}")

    entries = data_dict[split]

    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor()
    ])

    dataset = CAFOJsonDataset(entries, mode=task, transform=transform, image_root=image_root, infra_csv_path=infra_csv_path)
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)
    return loader
```

```python
import os
import json
import pandas as pd
import numpy as np
import ast
from datasets import GeneratorBasedBuilder, DatasetInfo, SplitGenerator, Split, Features, Value, Image, BuilderConfig

# Mapping dataset type keywords to filenames
DATASET_FILES = {
    'verified':  'splits/cafosat_verified_training.json',
    'augmented': 'splits/cafosat_augmented_training.json',
    'all':       'splits/cafosat_all_training.json',
    'set1':      'splits/cafosat_training_set1.json',
    'set2':      'splits/cafosat_training_set2.json',
    'merged':    'splits/cafosat_merged_training.json'
}

def parse_flag(val):
    if pd.isna(val):
        return 0.0
    if isinstance(val, str):
        val = val.strip().lower()
        if val in ['yes', 'true']:
            return 1.0
        try:
            return 1.0 if float(val) != 0 else 0.0
        except ValueError:
            return 0.0
    if isinstance(val, (int, float)):
        return 1.0 if val != 0 else 0.0
    return 0.0

class CAFOConfig(BuilderConfig):
    def __init__(self, name, file_key, mode='binary', **kwargs):
        super().__init__(name=name, **kwargs)
        self.file_key = file_key
        self.mode = mode  # 'binary', 'bbox', 'infra', or 'all'

class Builder(GeneratorBasedBuilder):
    DEFAULT_CONFIG_NAME = "set1_all"

    # Generate all config combinations
    BUILDER_CONFIGS = [
        CAFOConfig(
            name=f"{key}_{mode}",
            file_key=value,
            mode=mode,
            description=f"{key} split with {mode} mode"
        )
        for key, value in DATASET_FILES.items()
        for mode in ['binary', 'bbox', 'infra', 'all']
    ]

    def _info(self):
        return DatasetInfo(
            description="CAFO-SAT dataset: Satellite image patches with CAFO classification, infrastructure labels, and optional bounding boxes.",
            features=Features({
                "image": Image(),
                "label": Value("int32"),
                "infra": [Value("float32")],
                "bbox": [Value("float32")]
            })
        )

    def _split_generators(self, dl_manager):
        json_path = self.config.file_key
        with open(json_path, "r") as f:
            split_data = json.load(f)

        return [
            SplitGenerator(name=Split.TRAIN, gen_kwargs={"entries": split_data["train"]}),
            SplitGenerator(name=Split.TEST, gen_kwargs={"entries": split_data["test"]})
        ]

    def _generate_examples(self, entries):
        mode = self.config.mode
        infra_path = "infra_labels.csv"
        infra_data = pd.read_csv(infra_path).set_index("patch_file") if os.path.exists(infra_path) else None

        for idx, entry in enumerate(entries):
            img_path = os.path.join("data", entry["image_path"])
            label = int(entry.get("label", 0))
            output = {
                "image": img_path,
                "label": int(label > 0),  # Binary by default
                "infra": [0.0, 0.0, 0.0, 0.0],
                "bbox": [5.0, 5.0, 220.0, 220.0]
            }

            patch_file = os.path.basename(entry["image_path"])
            if infra_data is not None and patch_file in infra_data.index:
                row = infra_data.loc[patch_file]

                if mode in ['infra', 'all']:
                    output["infra"] = [
                        parse_flag(row.get("barn")),
                        parse_flag(row.get("manure_pond")),
                        parse_flag(row.get("grazing_area")),
                        parse_flag(row.get("others"))
                    ]

                if mode in ['bbox', 'all'] and int(label) > 0:
                    bbox = row.get("geom_bbox")
                    if isinstance(bbox, str):
                        try:
                            bbox_coords = list(ast.literal_eval(bbox))
                            scale = 224.0 / 833.0  # Resize scale
                            output["bbox"] = [coord * scale for coord in bbox_coords]
                        except Exception:
                            pass

            yield idx, output
```

```python
import pandas as pd
import numpy as np
import os
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import torchvision.transforms as transforms

def parse_flag(val):
    if pd.isna(val):
        return 0.0
    if isinstance(val, str):
        val = val.strip().lower()
        if val in ['yes', 'true']:
            return 1.0
        try:
            return 1.0 if float(val) != 0 else 0.0
        except ValueError:
            return 0.0
    if isinstance(val, (int, float)):
        return 1.0 if val != 0 else 0.0
    return 0.0

class CAFOCsvDataset(Dataset):
    def __init__(self, df, mode='binary', transform=None):
        self.df = df.reset_index(drop=True)
        self.mode = mode
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image_path = '/project/biocomplexity/gza5dr/CAFO/exp_v2/main_experiments/data_preparation/datas/'+row['patch_file']
        image = Image.open(image_path).convert("RGB")
        if self.transform:
            image = self.transform(image)

        label = int(row.get("label", 0))
        if self.mode == 'binary':
            label = int(label > 0)

        if self.mode == 'bbox' or self.mode == 'all':
            label = int(row.get("label", 0))

            if label > 0:
                bbox_val = row.get("geom_bbox", "[5.0, 5.0, 220.0, 220.0]")
                if isinstance(bbox_val, str):
                    try:
                        import ast
                        bbox = np.array(ast.literal_eval(bbox_val), dtype=np.float32)

                        # Scale from original image size to transformed size
                        original_size = 833
                        resized_size = image.size[0]  # .size returns (W, H)
                        scale = resized_size / original_size

                        bbox = np.array([coord * scale for coord in bbox], dtype=np.float32)
                    except:
                        bbox = np.array([5.0, 5.0, 220.0, 220.0], dtype=np.float32)
                else:
                    bbox = np.array([5.0, 5.0, 220.0, 220.0], dtype=np.float32)
            else:
                bbox = np.array([5.0, 5.0, 220.0, 220.0], dtype=np.float32)


        if self.mode == 'infra' or self.mode == 'all':
            infra = np.array([
               row.get("barn"),
               row.get("manure_pond"),
                row.get("grazing_area"),
               row.get("others")
            ], dtype=np.float32)

        if self.mode == 'bbox':
            return image, label, bbox
        elif self.mode == 'infra':
            return image, label, infra
        elif self.mode == 'all':
            return image, label, infra, bbox
        else:
            return image, label
```

```python
def get_cafo_dataloader_from_csv(csv_path, dataset_name='set1', split='train', task='binary',
                                  batch_size=32, shuffle=True, transform=None):
    """
    Args:
        csv_path: Path to the full metadata CSV
        dataset_name: one of set1, set2, merged, verified, etc.
        split: one of 'train', 'test', 'val'
        task: 'binary', 'bbox', 'infra', 'all'
    """
    df = pd.read_csv(csv_path)

    split_col = f"cafosat_{dataset_name}_training_{split}"
    if split_col not in df.columns:
        raise ValueError(f"Split column '{split_col}' not found in CSV.")

    # Filter based on split membership
    df_split = df[df[split_col] == 1]

    if transform is None:
        transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor()
        ])

    dataset = CAFOCsvDataset(df_split, mode=task, transform=transform)
    return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)
```

```python
loader = get_cafo_dataloader_from_csv(
    csv_path="/project/biocomplexity/gza5dr/CAFO/exp_v2/main_experiments/CAFOSat/datas/all_metadata/train_test_split.csv",
    dataset_name="verified",
    split="train",
    task="all",
    batch_size=16
)

for data in loader:
    images, labels, infra, bboxes = data
    print(images.shape, labels, infra.shape, bboxes.shape)
    break
```

```python
def show_image_with_bbox(image_tensor, bbox, infra_tensor):
    # Convert image tensor to PIL for plotting
    image = F.to_pil_image(image_tensor)

    # Unpack bbox
    x_min, y_min, x_max, y_max = bbox.tolist()

    fig, ax = plt.subplots(figsize=(6, 6))
    ax.imshow(image)
    ax.add_patch(plt.Rectangle(
        (x_min, y_min),
        x_max - x_min,
        y_max - y_min,
        edgecolor='red',
        facecolor='none',
        linewidth=2
    ))

    # Show infra flags
    labels = ["barn", "manure_pond", "grazing_area", "others"]
    flags = [int(x) for x in infra_tensor.tolist()]
    flag_str = "\n".join([f"{k}: {v}" for k, v in zip(labels, flags)])

    plt.title("Bounding Box + Infra\n" + flag_str)
    plt.axis('off')
    plt.show()

# Call the visualizer
show_image_with_bbox(image, bbox, infra_info)
```

```python
import pandas as pd
import numpy as np
import os
from torch.utils.data import Dataset
import torch

class CAFOCsvDataset(Dataset):
    def __init__(self, df, mode='binary', transform=None, load_from_npy=True):
        self.df = df.reset_index(drop=True)
        self.mode = mode
        self.transform = transform
        self.load_from_npy = load_from_npy

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        if self.load_from_npy:
            npy_path = row['patch_npy']
            image = np.load(npy_path)  # shape: (833, 833, 4)
            image = torch.tensor(image).permute(2, 0, 1).float() / 255.0  # to [C, H, W]
            if self.transform:
                image = self.transform(image)
        else:
            image_path = '/project/biocomplexity/gza5dr/CAFO/exp_v2/main_experiments/data_preparation/datas/' + row['patch_file']
            image = Image.open(image_path).convert("RGB")
            if self.transform:
                image = self.transform(image)

        label = int(row.get("label", 0))
        if self.mode == 'binary':
            label = int(label > 0)

        if self.mode in ['bbox', 'all']:
            if label > 0:
                bbox_val = row.get("geom_bbox", "[5.0, 5.0, 220.0, 220.0]")
                try:
                    import ast
                    bbox = np.array(ast.literal_eval(bbox_val), dtype=np.float32)

                    original_size = 833
                    resized_size = image.shape[1]  # torch tensor [C, H, W], so W
                    scale = resized_size / original_size
                    bbox = np.array([coord * scale for coord in bbox], dtype=np.float32)
                except:
                    bbox = np.array([5.0, 5.0, 220.0, 220.0], dtype=np.float32)
            else:
                bbox = np.array([5.0, 5.0, 220.0, 220.0], dtype=np.float32)

        if self.mode in ['infra', 'all']:
            infra = np.array([
                row.get("barn"),
                row.get("manure_pond"),
                row.get("grazing_area"),
                row.get("others")
            ], dtype=np.float32)

        if self.mode == 'bbox':
            return image, label, bbox
        elif self.mode == 'infra':
            return image, label, infra
        elif self.mode == 'all':
            return image, label, infra, bbox
        else:
            return image, label
```
